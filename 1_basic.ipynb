# **AMENITIES**

```python
mapping = {
    25: "Waste Disposal",
    26: "Rain Water Harvesting",
    28: "Bank Attached Property",
    2: "Power Back-up",
    5: "Feng Shui / Vaastu Compliant",
    8: "Private Garden / Terrace",
    10: "Centrally Air Conditioned",
    17: "Security / Fire Alarm",
    20: "Intercom Facility",
    24: "Water Storage",
    29: "Piped-gas",
    32: "Water purifier",
    39: "Near Bank",
    1: "Swimming Pool",
    3: "Club house / Community Center",
    6: "Park",
    9: "Security Personnel",
    12: "Fitness Centre / GYM",
    19: "Visitor Parking",
    21: "Lift(s)",
    23: "Maintenance Staff",
    33: "Shopping Centre",
    34: "WheelChair Accessibility",
    35: "DG Availability",
    36: "CCTV Surveillance",
    37: "Grade A Building",
    38: "Grocery Shop",
    11: "ATM",
    13: "Cafeteria / Food Court",
    15: "Bar / Lounge",
    16: "Conference room",
    22: "Service / Goods Lift",
    27: "Access to High Speed Internet"
}
```

```python
def clean_features(x):
    if pd.isna(x):
        return []

    nums = []
    for i in str(x).split(','):
        i = i.strip()
        if i.isdigit():
            i = int(i)
            if i in mapping:
                nums.append(mapping[i])
    return nums

df['FEATURES_LIST'] = df['AMENITIES'].apply(clean_features)
```

```python
from sklearn.preprocessing import MultiLabelBinarizer

mlb = MultiLabelBinarizer()

features_encoded = pd.DataFrame(
    mlb.fit_transform(df['FEATURES_LIST']),
    columns=mlb.classes_,
    index=df.index
)
```

```python
df.drop(columns=['AMENITIES', 'FEATURES_LIST'], inplace=True)
```
