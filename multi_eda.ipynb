```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
plt.figure(figsize=(15,6))
sns.heatmap(pd.crosstab(df['PROPERTY_TYPE'],df['LOCALITY_WO_CITY']))
```

```python
avg_price = df.groupby('LOCALITY_WO_CITY')['PRICE'].mean()
plt.figure(figsize=(5, 26))
sns.heatmap(avg_price.to_frame(), annot=True, cmap='YlGnBu')
plt.show()
```

```python
avg_price = df.groupby('LOCALITY_WO_CITY')['PRICE_PER_UNIT_AREA'].mean()
plt.figure(figsize=(5, 26))
sns.heatmap(avg_price.to_frame(), annot=True, cmap='YlGnBu')
plt.show()
```
